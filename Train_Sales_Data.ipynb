```python
import pandas as pd
```

```python
df = pd.read_csv("/content/train-chennai-sale.csv")
df
```

```python
df.columns

# lot of spelling mistakes in colums name
```

```python
df['AREA']= df['AREA'].replace({"Ana Nagar":"Anna Nagar",
                                "Ann Nagar":"Anna Nagar",
                                "Karapakam":"Karapakkam",
                                "Adyr":"Adyar",
                                "TNagar":"T Nagar",
                                "Velchery":"Velachery",
                                "Chrompt":"Chrompet",
                                "Chormpet":"Chrompet",
                                "Chrmpet":"Chrompet",
                                "KKNagar":"KK Nagar"})
```

```python
df['AREA'].unique()
```

```python
df['SALE_COND'].unique()
```

```python
df['SALE_COND']= df['SALE_COND'].replace({"Ab Normal":"AbNormal",
                                          "Partiall":"Partial",
                                          "PartiaLl":"Partial",
                                          "Adj Land":"AdjLand"})
```

```python
df['SALE_COND'].unique()
```

```python
df['PARK_FACIL'].unique()
```

```python
df["PARK_FACIL"] = df["PARK_FACIL"].apply(lambda i:"No" if i=="Noo" else i)
```

```python
df['PARK_FACIL'].unique()
```

```python
df["BUILDTYPE"].unique()
```

```python
df['BUILDTYPE']= df['BUILDTYPE'].replace({"Other":"Others",
                                          "Comercial":"Commercial"})
```

```python
df["BUILDTYPE"].unique()
```

```python
df["UTILITY_AVAIL"].unique()
```

```python
df['UTILITY_AVAIL']= df['UTILITY_AVAIL'].replace({"AllPub":"All Pub",
                                          "NoSeWa":"NoSewr "})
```

```python
df["UTILITY_AVAIL"].unique()
```

```python
df["STREET"].unique()
```

```python
df['STREET']= df['STREET'].replace({"Pavd":"Paved",
                                          "NoAccess":"No Access"})
```
